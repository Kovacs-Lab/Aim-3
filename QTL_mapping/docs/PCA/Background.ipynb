This file will attempt to provide some background on PCA, and describe what it is doing. I am no expert in linear algebra, so inaccuracies may exist in this explanation. Principal component analysis is a way to describe a portion of the variance expressed in x dimensions in two or so dimensions. This allows points that exist in x dimensional space to be plotted in a meaningful way in one or more two dimensional plots. 

```r
plot(5)
```

This point has no variance, as it is one point.

```r
plot(c(1,2,3,4,5),c(1,2,3,4,5))
```

This plot has variance in one dimension. The same information could be expressed on a number line. 

```r
x <- c(1,2,3,4,5)
stripchart(x)
```

See? No information is lost in only 1 dimension. 

```r
plot(c(1.25,1,1.75,2,2.5,3,3.5,4,4.5,5),c(1.75,1,1.25,2,3.5,3,4,4,4.75,5))
eq <- function(x){x}
eq1 <- function(x){-1*x+6}
curve(eq,from = 1, to = 5, add = TRUE, col = "blue")
curve(eq1,from = 2.5, to = 3.5, add = TRUE, col = "red")
```

This set of data now has variation on two axes, which is easily interpreted in two dimensions. This variation occurs more in one direction (the blue direction) than the other (the red direction). This means that the blue direction would contribute more to the overall variation of the dataset. This is how the primary compenents are calculated. These directions of variance can be expressed as vectors. Before this conversion to vectors, the data is centered and scaled such that the magnitudes of the vectors are comparable. Each one of these vectors, or directions of variance has a few notable properties. Firstly, all of these directions of variance are orthoganl, or perpendicular. So, in two dimensional space, it is just like the x and y axes. In three dimensions the directions of variance would be related to each other in just the same way as the x, y and z axes. This gets harder and harder to visualize in higher dimensions, which is why PCA is used. Next, these vectors' direction and magnitude are determined by the individual points, called observations, that vary in that direction. Many points that are more spread apart will lead to a vector with a higher magnitude, and fewer points that are closer together will lead to a lower magnitude. Furthermore, these vectors can be "combined" to make a vector with a lower dimensionality that expresses much of the variation present in a given dataset. 

```r
plot(c(1.25,1,1.75,2,2.5,3,3.5,4,4.5,5),c(1.75,1,1.25,2,3.5,3,4,4,4.75,5))
eq <- function(x){x}
eq1 <- function(x){-1*x+6}
curve(eq,from = 1, to = 5, add = TRUE, col = "blue")
curve(eq1,from = 2.5, to = 3.5, add = TRUE, col = "red")
segments(x0 = 3, y0 = 1.5, x1 = 3, y1 = 4.5, col = 'green')
```

This combonation is shown in green, and it is more helpful to think of this combonation not as a vector moving throught the plane defined by two directions of variance, but rather a vector moving through two directions independantly, as in hihgher dimensional data sets this is the case. For example, if this data set showed variance in 20 dimensions each one of the combined vectors would move through all 20 of those directions. Each one of these vectors is called an eigenvector, and while this example made it easy to graphically show which way that vector is going, during PCA this is found via decomoposition. Depending on how the dataset is arranged with will either be eigendecomposition or single value decomposition. These eigenvectors all have an associated eigenvalue, which in the context of PCA is basicaly how "principal" each component is. Lastly, the observations are then plotted on the highest eigenvalue eigenvectors.
