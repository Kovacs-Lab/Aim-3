Which points do we use to determine length of veins? Used average instead

```r
#Using pythagorean distance formula (d=√(x2-x1)2+(y2-y1)2) to find length of:
#midvein
midvein4 <- (((df$X21-df$X4)^2) + ((df$Y21-df$Y4)^2))^0.5
midvein3 <- (((df$X21-df$X3)^2) + ((df$Y21-df$Y3)^2))^0.5
midvein.ave <- (midvein4 + midvein3)/2
#petiolar junction to distal sinus
distal.sinus <- (((df$X19-df$X3)^2) + ((df$Y19-df$Y3)^2))^0.5
#petiolar junction to distal lobe tip
distal.lobe3 <- (((df$X18-df$X3)^2) + ((df$Y18-df$Y3)^2))^0.5
distal.lobe2 <- (((df$X18-df$X2)^2) + ((df$Y18-df$Y2)^2))^0.5
distal.ave <- (distal.lobe3 + distal.lobe2)/2
#petiolar junction to proximal sinus
proximal.sinus <- (((df$X16-df$X2)^2) + ((df$Y16-df$Y2)^2))^0.5
#petiolar junction to proximal lobe tip
proximal.tip2 <- (((df$X15-df$X2)^2) + ((df$Y15-df$Y2)^2))^0.5
proximal.tip1 <- (((df$X15-df$X1)^2) + ((df$Y15-df$Y1)^2))^0.5
proximal.ave <- (proximal.tip2 + proximal.tip1)/2
#petiolar vein
petiolar.vein6 <- (((df$X14-df$X6)^2) + ((df$Y14-df$Y6)^2))^0.5
petiolar.vein5 <- (((df$X14-df$X5)^2) + ((df$Y14-df$Y5)^2))^0.5
petiolar.ave <- (petiolar.vein6 + petiolar.vein5)/2

#output results of length calculations
head(cbind(df$Label, midvein4, midvein3, midvein.ave), 34)
head(cbind(df$Label, distal.lobe3, distal.lobe2, distal.ave), 34)
head(cbind(df$Label, proximal.tip2, proximal.tip1, proximal.ave), 34)
head(cbind(df$Label, petiolar.vein1, petiolar.vein2, petiolar.ave), 34)
head(cbind(df$Label, distal.sinus, proximal.sinus), 34)
```

Need to determine which angle between proximal and petiolar vein; (14-5-15)? (14-1-15)? (14-1-6)?

```r
#Using Law of Cosines (b2=a2+c2-2ac*cos(B))to find angles between:
#midrib and distal vein
mid.distal.b <- (((df$X21-df$X18)^2) + ((df$Y21-df$Y18)^2))^0.5
cos.mid.distal <- ((midvein.ave^2 + distal.ave^2 - mid.distal.b^2)/(2*midvein.ave*distal.ave))
mid.distal <- (acos(cos.mid.distal))*180/pi
#distal and proximal vein
distal.prox.b <- (((df$X18-df$X15)^2) + ((df$Y18-df$Y15)^2))^0.5
cos.distal.prox <- ((distal.ave^2 + proximal.ave^2 - distal.prox.b^2)/(2*distal.ave*proximal.ave))
distal.prox <- (acos(cos.distal.prox))*180/pi
#proximal and petiolar vein
prox.petiolar.b <- (((df$X15-df$X14)^2) + ((df$Y15-df$Y14)^2))^0.5
cos.prox.petiolar <- ((proximal.ave^2 + petiolar.ave^2 - prox.petiolar.b^2)/(2*proximal.ave*petiolar.ave))
prox.petiolar <- (acos(cos.prox.petiolar))*180/pi

head(cbind(df$Label, mid.distal, distal.prox, prox.petiolar), 34)
```
