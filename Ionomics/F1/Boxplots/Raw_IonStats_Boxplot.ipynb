```python
import numpy as np #It provides a high-performance multidimensional array object, and tools for working with these arrays, Array creation, array indexing, 
import pandas as pd #for data manipulation and analysis, Reshaping and pivoting of data sets, Data structure column insertion and deletion, Data set merging and joining.
import matplotlib.pyplot as plt #used to create line plot, histograms, scatter plots,line plot, 3-D plot etc.
import scipy.stats as stats #Easy to use and understand as well as fast computational power.
```

```python
data0 = np.genfromtxt('Raw Ionomics Data_F1 pop.csv',delimiter = ',', skip_header = 2) #genfromtext imports raw ionomics data of F1 population and convert it into appropriate data type and it read aithout the 2nd header
data0 = pd.DataFrame(data0) #a two dimentional datasheet to illustrate the data is created(like a excelsheet)
data = data0.drop(data0.columns[0],axis = 1) #the marking on first column says "0" for counting sample number, so it just deletes it.
plt.rcParams['figure.figsize'] = (20,10) #width, height in inches (this is the default size of the figures, the figures we are gonna make will automatically make figures in this dimention unless stated so)
```

```python
plt.figure(figsize=(5,50)) #as stated erlier this is also the figure size of the graph but it has different size 5,50 not accouding to standard figure size mentioned above.
plt.matshow(data,fignum =1) #this represents the figure just below and it plots the abosrption rate of diffrent minerals, the line on the middle is that of calcium and it is way higher than all the other elements. The numbers on the top 0-20 shows the number assigned to each minerals as stated in the csv file, for eg. calcium is in 8th number.
```

```python
plt.matshow(data.corr()) #this figure below just shows the correlation between different elements. we have minerals on both axis and we look how much correletion is between these elements. the line on the middle is yellow because each element is highly correlated to itself. eg lithium on the first line is highly correlated to aluminium which is on the 5th line.
```
