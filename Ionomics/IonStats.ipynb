```python
import numpy as np #this allows you to import the module with a shortened name np and work with all the functions present
import pandas as pd #Pandas is a python data analysis library which takes data and creates a python object with rows and columns called data frame that makes it easier to work with comparison, lists, or comprehension of data.
import matplotlib.pyplot as plt #matplotlib is a libray in python that allows for data visualization.
import scipy.stats as stats #this is a module that contains a large number of probability distributions as well as a library of statistical functions. 
```

```python
data = np.genfromtxt('New_ionStats.csv',delimiter = ',') #genfromtext takes missing data into account.
data = pd.DataFrame(data) #DataFrame is a two dimensional labeled structure data with columns of different types that is similar to a spreadsheet.
plt.rcParams['figure.figsize'] = (20,10) #This allows you to chage the size of figures drawn.
```

They skewness and kurtosis of this data set would suggest that we use a non- normal based analysis, remove outliers, or transform the data to reduce skewness and kurtosis. 

```python
ion = data[3] 
print(ion[1]) #to double check what col you're on
mu = ion.mean() #Calculating the average value for our data
sigma = ion.std() #sigma is the greek letter for standard deviation so we are calculating the standard deviation using .std
x1 = ion.quantile(.25) #the 25th percent quartile corresponds to the first of three ranges that make up a distribution
x2 = ion.quantile(.75) #The 75th perccintile is the third quartile range 
x_all = np.arange(ion.min(), ion.max(), 0.001) #interquartile range is the range between the first and third quartile and it makes up the middle 50 percent of a distribution and and is useful for skewed distributions beacuse the interquartile range is not affected by the presence of a few extreme scores

z1 = ( x1 - mu ) / sigma #Calculation for finding the Z score 
z2 = ( x2 - mu ) / sigma
x = np.arange(z1, z2, 0.001)
y = stats.norm.pdf(x,0,1) #checks the accuracy of our numbers
y2 = stats.norm.pdf(x_all,0,1)

fig ,ax = plt.subplots(figsize=(9,6)) #this unpacks a tupule into variables fig and ax in a more concise way that will allow you to alter your plots.
ax.hist(ion,alpha = .5) #Plots a histogram graph.
ax.plot(x_all,y2*150) #Plots the axes and gives you control to customize your plot space. 
ax.fill_between(x,y*150,0, alpha=0.3, color='b') #Fills in the area between two horizontal curves.
ax.set_xlim([ion.min()-1,ion.max()+1]) #This allows you to focus on a specific range of x values.
ax.set_xlabel('# of Standard Deviations Outside the Mean') #Setting an x label to the axes.
ax.set_title("B11",fontsize = 24) #this allows you to a title to subplots in matplotlib
print("Mean: ",mu,"\nStandard Deviation: ",sigma,"\n25th Percentile (Perdicted): ",x1,
      "\n50th Percentile (Predicted, also the median): ",ion.quantile(.50),"\n75th Percentile (perdicted): ",x2,
     "\nInterquartile Width (perdicted): ",x2-x1,"\n Minimum: ",ion.min(),"\n Maximum: ",ion.max(),
     "\nVariance: ",ion.var(),"\nSkewness: ",ion.skew(),"\nKurtosis: ",ion.kurt())
```

The skewness and kurtosis of this data set would suggest that we can likely use normal based analysis without removal of outliers or transformation.
