```python
data0 = np.genfromtxt('geneva_F1_combined_ionomics_raw_Z.csv',delimiter = ',', skip_header = 1)
data0 = pd.DataFrame(data0)
data = data0.drop(data0.columns[0:2],axis = 1)
plt.rcParams['figure.figsize'] = (20,10)
```

```python
z = pd.DataFrame(np.arange(8778).reshape(418,21))
z = z.astype('float32')
z.drop(z.columns[0:2],axis = 1, inplace = True )
z
```

```python
for x in range(2,(data.shape[1]+2)):
    mean = data[x].mean()
    stdev = data[x].std()
    print(x,"m: ",mean,"s: ",stdev)
    for y in range(data.shape[0]):
        zscore = (data[x][y] - mean)/stdev
        z[x][y] = zscore
        if data[x][y] < 0:
            z[x][y] = "NaN"
z
```

```python
plt.figure(figsize=(5,50))
plt.matshow(z,fignum =1)
```

```python
for x in range(2,(z.shape[1]+2)): # counts through columns
    print(x,"k: ", z[x].kurt(),"s: ", z[x].skew(),"v: ",z[x].var())
    if z[x].kurt() > 40:
        cutoff = z[x].quantile(.99)
        for y in range(z.shape[0]): #counts through rows
            if z[x][y] > cutoff:
                z[x][y] = 'NaN'
z
```
