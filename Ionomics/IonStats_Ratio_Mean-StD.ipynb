```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
data0 = np.genfromtxt('New_ionStats_Ratio.csv',delimiter = ',', skip_header = 1)
data0 = pd.DataFrame(data0)
data = data0.drop(data0.columns[0:2],axis=1)
plt.rcParams['figure.figsize'] = (20,10)
```

```python
data
```

```python
g1count=0 #gx count means group x count
g2count=0
g3count=0
g4count=0
g5count=0
g6count=0
g7count=0
g8count=0
g9count=0
for x in range(len(data[2])):
    if (data[2][x]==1):
        g1count+=1
    elif (data[2][x]==2):
        g2count+=1
    elif (data[2][x]==3):
        g3count+=1
    elif (data[2][x]==4):
        g4count+=1
    elif (data[2][x]==5):
        g5count+=1
    elif (data[2][x]==6):
        g6count+=1
    elif (data[2][x]==7):
        g7count+=1
    elif (data[2][x]==8):
        g8count+=1
    elif (data[2][x]==9):
        g9count+=1
```

```python
print(g1count,g2count,g3count,g4count,g5count,g6count,g7count,g8count,g9count)
```

```python
g1 = data[0:g1count]
g2 = data[g1count:g1count+g2count]
g3 = data[g1count+g2count:g1count+g2count+g3count]
g4 = data[g1count+g2count+g3count:g1count+g2count+g3count+g4count]
g5 = data[g1count+g2count+g3count+g4count:g1count+g2count+g3count+g4count+g5count]
g6 = data[g1count+g2count+g3count+g4count+g5count:g1count+g2count+g3count+g4count+g5count+g6count]
g7 = data[g1count+g2count+g3count+g4count+g5count+g6count:g1count+g2count+g3count+g4count+g5count+g6count+g7count]
g8 = data[g1count+g2count+g3count+g4count+g5count+g6count+g7count:g1count+g2count+g3count+g4count+g5count+g6count+g7count+g8count]
g9 = data[g1count+g2count+g3count+g4count+g5count+g6count+g7count+g8count:g1count+g2count+g3count+g4count+g5count+g6count+g7count+g8count+g9count]
```

```python
for x in range(len(g1)): #mgx is the mean of group x #stdx is the standard deviation of group x
    mg1 = g1.mean()
    std1 = g1.std()
for x in range(len(g2)):
    mg2 = g2.mean()
    std2 = g2.std()
for x in range(len(g3)):
    mg3 = g3.mean()
    std3 = g3.std()
for x in range(len(g4)):
    mg4 = g4.mean()
    std4 = g4.std()
for x in range(len(g5)):
    mg5 = g5.mean()
    std5 = g5.std()
for x in range(len(g6)):
    mg6 = g6.mean()
    std6 = g6.std()
for x in range(len(g7)):
    mg7 = g7.mean()
    std7 = g7.std()
for x in range(len(g8)):
    mg8 = g8.mean()
    std8 = g8.std()
for x in range(len(g9)):
    mg9 = g9.mean()
    std9 = g9.std()
```

```python
stddf=pd.DataFrame()
pd.DataFrame().shape
stddf=stddf.drop(stddf.columns[2])
stddf[0] = std1
stddf[1] = std2
stddf[2] = std3
stddf[3] = std4
stddf[4] = std5
stddf[5] = std6
stddf[6] = std7
stddf[7] = std8
stddf[8] = std9
stddf = stddf.transpose()
plt.matshow(stddf)
stddf
```
