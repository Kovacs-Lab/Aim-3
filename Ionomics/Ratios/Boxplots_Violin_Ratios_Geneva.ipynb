```r
Elements <- read.csv("New_ionStats_Ratio_Geneva.csv")
```

```r
boxplot(Elements$B11.P31~Elements$Spec_int) #Boron #from here on .P31 signifies ratio, colon not used since it is an operator in R
Elements$Spec_int <- as.factor(Elements$Spec_int)
boron <- ggplot(Elements, aes(x=Spec_int, y=B11.P31)) + 
  geom_violin()
boron
```
