```r
Elements <- read.csv("New_ionStats_Ratio.csv")
```

```r
boxplot(Elements$B11~Elements$Spec_int) #Boron
Elements$Spec_int <- as.factor(Elements$Spec_int)
boron <- ggplot(Elements, aes(x=Elements$Spec_int, y=Elements$B11)) + 
  geom_violin()
boron
```
